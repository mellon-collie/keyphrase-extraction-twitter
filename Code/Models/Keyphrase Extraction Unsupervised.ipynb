```python
import os
os.chdir("hulth2003/TestingZip/Testing/")
```

```python
filenames = [x for x in os.listdir() if x.endswith('.abstr') and os.path.getsize(x) > 0]
```

```python
data = []
for filename in filenames:
    file = open(filename, "r")
    data.append(file.read())
```

```python
# initialize keyphrase extraction model
#all the unsupervised models listed below

all_keyphrases = []
extractor = pke.unsupervised.YAKE()
# extractor = pke.unsupervised.TfIdf()
# extractor = pke.unsupervised.KPMiner()
# extractor = pke.unsupervised.TextRank()
# extractor = pke.unsupervised.SingleRank()
# extractor = pke.unsupervised.TopicRank()
# extractor = pke.unsupervised.TopicalPageRank()
# extractor = pke.unsupervised.PositionRank()
# extractor = pke.unsupervised.MultipartiteRank()

# load the content of the document, here document is expected to be in raw
# format (i.e. a simple text file) and preprocessing is carried out using spacy

def find_key_words(ip_string, n_gram_size, num_keyword):
    extractor.load_document(input= ip_string, language='en')

    # keyphrase candidate selection, in the case of TopicRank: sequences of nouns
    # and adjectives (i.e. `(Noun|Adj)*`)
    extractor.candidate_selection(n = n_gram_size)

    # candidate weighting, in the case of TopicRank: using a random walk algorithm
    extractor.candidate_weighting()

    # N-best selection, keyphrases contains the 10 highest scored candidates as
    # (keyphrase, score) tuples
    keyphrases = extractor.get_n_best(n = num_keyword)
    all_keyphrases.append(keyphrases)
```

```python
for doc in data:
    length = len(doc)
    find_key_words(doc, 1, 5)
```
